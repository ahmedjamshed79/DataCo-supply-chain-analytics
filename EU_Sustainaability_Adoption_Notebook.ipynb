```python
# Ensure data is sorted by country and year
data = df.sort_values(by=['country', 'year'])

# Calculate the GDP Growth % using the pct_change() method
data['GDP Growth %'] = data.groupby('country')['gdp'].pct_change() * 100

# Display the dataset to verify the new column
print(data[(data['country'] == 'United Kingdom') & (data['year'] == 2017)][['country', 'year', 'gdp', 'GDP Growth %']].head())
```
